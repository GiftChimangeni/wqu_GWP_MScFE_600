```python
# TODO: @GiftRyanChmangeni: EDA for regulation risks asociated with this lending scenario

# Define the start and end dates for fetching data
end_date = datetime.datetime(2024, 7, 1)
start_date = end_date - datetime.timedelta(365 * 5)  # 3 years & 3 months of data from FRED

# Define the ticker symbol for NDR12MCD (National Deposit Rates: 12-Month CD)
cds_ticker = 'NDR12MCD'

# Fetch data from FRED
try:
    df_cds_spreads = web.DataReader(cds_ticker, 'fred', start_date, end_date)
    print("Successfully fetched CDS Spreads Data:")
    print(df_cds_spreads.head())

    # Basic Statistics
    print("\nSummary Statistics:")
    print(df_cds_spreads.describe())

    # Kurtosis and Skewness
    print("\nKurtosis:")
    print(kurtosis(df_cds_spreads['NDR12MCD']))

    print("\nSkewness:")
    print(skew(df_cds_spreads['NDR12MCD']))

    # Time Series Visualization
    plt.figure(figsize=(12, 6))
    plt.plot(df_cds_spreads.index, df_cds_spreads['NDR12MCD'], marker='o', linestyle='-')
    plt.title('National Deposit Rates: 12-Month CD (NDR12MCD)')
    plt.xlabel('Date')
    plt.ylabel('Rate (%)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Error fetching data: {e}")
```

**Insights Based on the results of the National Deposit Rates: 12-Month CD (NDR12MCD) data:**

**Regulatory Compliance**: The data provides a detailed view of interest rate fluctuations, crucial for ensuring compliance with regulatory caps and guidelines governing unsecured lending.

**Lending Practices Impact**: With a mean rate of 0.937% and a standard deviation of 0.739%, institutions can adjust their fixed-rate lending strategies to align with market trends and regulatory requirements.

**Risk Assessment**: The negative kurtosis (-1.798) indicates a relatively flat distribution with fewer extreme values, suggesting reduced tail risk in interest rate movements. This helps in planning risk mitigation strategies for fixed-rate lending.

**Market Sensitivity**: Despite minimal skewness (0.052), which indicates a near-symmetrical distribution around the mean, institutions must monitor subtle shifts in NDR12MCD rates to stay responsive to market dynamics and competitive pressures.

**Comparative Analysis:** Benchmarking NDR12MCD against other financial indicators enables lenders to assess relative risk exposure and optimize lending decisions to maintain profitability and regulatory compliance.

```python
# TODO @GiftRyanChmangeni: EDA for regulation risks asociated with this lending scenario

# Define the ticker symbol for the publicly traded equity
ticker_symbol = 'AAPL'  # Apple Inc. ticker symbol

# Fetch data from Yahoo Finance
try:
    # Fetch historical data
    stock_data = yf.download(ticker_symbol, start='2010-01-01', end='2024-07-01')

    # Display summary statistics
    print("\nSummary Statistics:")
    print(stock_data['Adj Close'].describe())

    # Calculate Kurtosis and Skewness
    print("\nKurtosis:")
    print(stock_data['Adj Close'].kurtosis())

    print("\nSkewness:")
    print(stock_data['Adj Close'].skew())

    # Time Series Visualization
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data.index, stock_data['Adj Close'], marker='o', linestyle='-', color='darkblue', label=ticker_symbol)
    plt.title(f'Historical Stock Prices: {ticker_symbol}')
    plt.xlabel('Date')
    plt.ylabel('Price ($)')
    plt.grid(True)
    plt.tight_layout()
    plt.legend()
    plt.show()

    # Plotting volatility
    plt.figure(figsize=(12, 6))
    stock_data['Daily Returns'] = stock_data['Adj Close'].pct_change()
    sns.histplot(stock_data['Daily Returns'].dropna(), bins=100, kde=True, color='darkblue')
    plt.title(f'Daily Returns Distribution for {ticker_symbol}')
    plt.xlabel('Daily Returns')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Error fetching data: {e}")
```

**Apple Inc. Adjusted Close Price Analysis: Insights into Regulatory Compliance, Market Impact, and Risk Assessment**

**Regulatory Compliance:**
Analyzing adjusted closing prices of publicly traded equity is crucial for regulatory compliance. It helps regulators ensure financial stability and consumer protection by assessing lenders' risk exposure and adherence to regulatory guidelines.

**Lending Practices Impact:**
Fluctuations in adjusted closing prices directly influence lending practices. Higher volatility (59.38%) prompts lenders to adjust financing terms and rates to mitigate potential losses, based on market fluctuations and risk tolerance levels.

**Risk Assessment:**
The analysis provides insights into market risk associated with publicly traded equity. The significant variability (59.38% standard deviation) and distribution characteristics (skewness of 0.992, kurtosis of -0.579) inform risk management strategies and decisions on loan terms.

**Market Sensitivity:**
Adjusted closing prices reflect high sensitivity to market conditions and economic events. Understanding these fluctuations helps lenders optimize loan terms, collateral requirements, and interest rates to manage exposure effectively.

**Comparative Analysis:**
Comparing adjusted closing prices with benchmarks informs competitive lending strategies. It assists in benchmarking against industry standards, evaluating market sentiment, and making data-driven decisions in dynamic lending environments.

These insights empower lenders to navigate regulatory requirements and optimize lending decisions effectively.

**Insights on Understanding Risk in Below-Investment-Grade Corporate Bonds**

**Regulatory Compliance:**

Regulatory compliance here means following rules that limit how much risk financial institutions can take with these bonds. It's crucial for institutions to manage their exposure carefully and report it accurately.

**Lending Practices Impact:**
The higher the spread (the difference in yields between risky bonds and safe bonds), the more cautious lenders become about lending money. This can make borrowing more expensive and harder to get, as lenders try to protect themselves from the risks associated with these bonds.

**Risk Assessment:**
The data's average spread of about 4.23% with a standard deviation of 1.12% shows that risks can vary a lot over time. High kurtosis (6.32) means there are often big swings in spreads, indicating periods of high risk. Skewness (2.13) shows that large increases in spreads happen more often than big decreases. To manage these risks, institutions need strong methods to predict and handle sudden changes in the market.

**Market Sensitivity:**
The numbers in the data and the kurtosis and skewness values show how sensitive these bonds are to changes in the economy, interest rates, and investor confidence. This sensitivity means it's important to watch the market closely to avoid surprises that could harm investments.

**Comparative Analysis:**
Comparing this data with other bond indices or safer investments helps institutions decide where to invest money for the best balance of risk and return. By comparing, they can also make sure they're following regulations and managing risk effectively.

In summary, using the ICE BofA US High Yield Index Option-Adjusted Spread data helps financial institutions follow rules, adjust lending decisions, understand and manage risks, and make smart investment choices based on market conditions.
